```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("train.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.groupby('Survived')['PassengerId'].count()
```

```python
f, ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='Survived',y='Age',data=data)
```

```python
f, ax=plt.subplots(figsize=(7,7))
sns.barplot(x='Sex',y='Survived',data=data)
```

```python
f, ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='Sex',y='Age',data=data,hue='Survived')
```

```python
sns.barplot(x='Pclass',y='Survived',data=data)
```

```python
sns.barplot(x='Pclass',y='Survived',data=data,hue='Sex')
```

```python
sns.barplot(x='SibSp',y='Survived',data=data)
```

```python
sns.barplot(x='Parch',y='Survived',data=data)
```

```python
survived=data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
```

```python
survived=data.loc[data.Survived==0,'Age'].dropna()
sns.distplot(survived)
```

```python
sns.pairplot(data.dropna())
```

```python
grid=sns.FacetGrid(data,col='Survived',row='Pclass',size=2.4,aspect=1.5)
grid.map(plt.hist,'Age',alpha=.5, bins=20)
grid.add_legend()
```

```python
grid=sns.FacetGrid(data,row='Embarked',col='Survived',size=2.2,aspect=1.6)
grid.map(sns.barplot,'Sex','Fare',alpha=.5)
```

```python
grid=sns.FacetGrid(data,row='Pclass',col='Sex')
grid.map(plt.hist,'Age',Alpha=0.5,bins=50)
```

```python
sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.title("QueensTown")
plt.xticks([0,1])
plt.show()
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.title("Southampton")
plt.xticks([0,1])
```

```python
sns.heatmap(data.corr(),cmap='CMRmap')
plt.legend()
```

```python
figbi, axsbi=plt.subplots(2,4,figsize=(16,10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axsbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axsbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axsbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axsbi[0,3],xlim=[0,1])
```
